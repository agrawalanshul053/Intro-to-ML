```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MR data - Compiled Data Set.csv')
# df
```

```python
df = df.drop(['Title','Text'],axis=1)
# df
```

```python
df = df.fillna(0)
# df
```

```python
col = df.columns
# col
```

```python
df.columns = [i for i in range(0,len(col))]
# df.info
# df.head(30)
```

```python
dic={0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4, 'Personal Vendetta':5, 'Love Affairs':6, 'Casteism':7,'Unknown reasons':8}
```

```python
df[1]=df[1].apply(lambda x : dic[x])
```

```python
df[11]=df[12]=0
# df

for i in range(len(df)):
    if ',' in str(df.iloc[i,8]):
        lis = str(df.iloc[i,8]).split(',')
        df.iloc[i,8] = int(lis[0])
        df.iloc[i,11] = int(lis[1])
        if len(lis)>2:
            df.iloc[i,12] = int(lis[2])
    elif int(df.iloc[i,8])>100:
        df.iloc[i,8] = df.iloc[i,8]//100
        df.iloc[i,11] = df.iloc[i,8]%100
print(df)
df.describe()
```

```python
count1 = 0
count2 = 0
count3 = 0
count4 = 0
for i in range(len(df)):
    lis = list(df.iloc[i,[8,11,12]])
    if 7 in lis :
        count1 += 1
        if 8 in lis:
            count2 += 1
            if 10 in lis:
                count3 += 1
        if 10 in lis:
            count4 += 1
# print(count1, count2, count3, count4)

import numpy as np
y = np.array([count1-count2-count4+count3,count2,count4])
mylabels = ["Others", "Assault on Women with Intent to Outrage her Modesty(crime number 8)", "Protection of Children from Sexual Offences Act(Crime number 10)"]
%matplotlib inline
plt.pie(y, labels = mylabels)
# plt.legend(title = "Rape only:")
plt.title("Crime related to women rape (crime number 7)")
plt.show()
```

```python
lis = list(df[1].value_counts().index)
Number = list(df[1].value_counts())
# print(lis)
# df[1].value_counts()

x = np.array(lis)
plt.pie(Number[1:],labels = x[1:])
plt.title("Crime related to murder")
plt.legend(title = "Reasons:")
plt.show()
```
